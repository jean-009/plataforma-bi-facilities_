```python
import pandas as pd
from pathlib import Path
import os
import io # Usaremos io.StringIO se houver CSVs

# -------------------------------------------------------------------------------------------------
# 1. PARÂMETROS - AJUSTE A EXTENSÃO DOMINANTE AQUI
# -------------------------------------------------------------------------------------------------
CAMINHO_RAIZ = Path('../01_raw_data')
EXTENSAO = '.xlsx' # <-- Mude para '.xls' ou '.csv' se for o caso!

# -------------------------------------------------------------------------------------------------
# 2. FUNÇÃO DE EXTRAÇÃO E LIMPEZA
# -------------------------------------------------------------------------------------------------

def processar_planilhas_em_lote():
    """
    Varre a pasta de dados brutos, extrai o cabeçalho e as dimensões de cada arquivo.
    
    Esta função é apenas para ANÁLISE INICIAL. O código real de ETL será mais complexo.
    """
    arquivos_encontrados = list(CAMINHO_RAIZ.glob(f'*{EXTENSAO}'))
    
    if not arquivos_encontrados:
        print(f"❌ Erro: Nenhum arquivo com a extensão {EXTENSAO} encontrado em {CAMINHO_RAIZ}")
        print("   Verifique se a extensão está correta e se os arquivos estão na pasta.")
        return

    print(f"✅ Encontrados {len(arquivos_encontrados)} arquivos {EXTENSAO}.")
    print("\n--- ANÁLISE DAS PRIMEIRAS LINHAS ---")

    # Limitamos a 5 arquivos para não sobrecarregar seu notebook na primeira rodada
    for i, arquivo in enumerate(arquivos_encontrados[:5]): 
        print(f"\n[Arquivo {i+1} de {len(arquivos_encontrados)}]: {arquivo.name}")

        try:
            # Lógica para ler Excel (.xlsx / .xls)
            if EXTENSAO in ['.xlsx', '.xls']:
                # Tenta ler as 100 primeiras linhas, assumindo header=None
                df_amostra = pd.read_excel(arquivo, header=None, nrows=100)
            
            # Lógica para ler CSV (se a extensão for .csv)
            elif EXTENSAO == '.csv':
                 df_amostra = pd.read_csv(arquivo, header=None, nrows=100)
            
            else:
                print("   Extensão não suportada neste script de análise inicial.")
                continue

            # Exibe as 10 primeiras linhas e dimensões
            print(df_amostra.head(10).to_string()) 
            print(f"\n   Dimensões (Linhas x Colunas): {df_amostra.shape}")
            
        except Exception as e:
            print(f"   ⚠️ Falha ao ler o arquivo: {e}")

# -------------------------------------------------------------------------------------------------
# 3. EXECUÇÃO
# -------------------------------------------------------------------------------------------------

processar_planilhas_em_lote()
```
